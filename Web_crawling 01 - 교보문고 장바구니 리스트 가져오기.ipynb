#### Selenium 관한 간단한 함수들 

```python
# URL에 접근하는 api     # get(‘http://url.com’)

# 페이지의 단일 element에 접근하는 api,(처음 찾아지는 요소를 추출한다. )
# find_element_by_name(‘HTML_name’)                              # name 속성으로 요소 하나 추출
# find_element_by_id(‘HTML_id’)                                  # id 속성으로 요소 하나 추출
# find_element_by_xpath(‘/html/body/some/xpath’)                 # xpath를 지정해 요소 하나 추출 
# find_element_by_css_selector(‘#css > div.selector’)            # CSS 선택자로 요소 하나 추출        
# find_element_by_class_name(‘some_class_name’)                  # 클래스 이름이 name에 해당하는 요소 하나 추출
# find_element_by_tag_name(‘name’)                               # 태그 이름이 name에 해당하는 요소 하나 추출
# find_element_by_link_text('text')                              # 링크 텍스트로 요소 추출
# find_element_by_partial_link_text('text')                      # 링크의 자식 요소에 포함돼 있는 텍스트로 요소 하나 추출
# 위에서 'elements'를 해주면 모든 요소를 추출할 수 있다. 

# click()    요소를 클릭
# submit()   입력 양식 전송
```

## 3. 장바구니에서 리스트 가져오기

#### 로그인 이후 장바구니 페이지로 이동한다.

```python
web_crawling.get('http://order.kyobobook.co.kr/cart/cartListMain')
```

#### BeautifulSoup 패키지를 이용하여 장바구니에서 책 리스트의 정보를 가져온다.

```python
from bs4 import BeautifulSoup

# pip install beautifulsoup4
```

#### 장바구니 페이지 소스를 전부 가져오기

```python
html = web_crawling.page_source             
```

#### BeautifulSoup으로 파싱하기(parser:컴파일, 데이터 변환)

```python
soup = BeautifulSoup(html, 'html.parser')  
```

#### 책 제목이 있는 부분의 정보 가져와 리스트형태로 저장 

```python
cart = soup.select('div.book_name > a > span')
```

```python
img = imread('start_sel24.jpg')
plt.figure(figsize = (30, 30))
plt.imshow(img)
plt.show()
```

#### 리스트 형태에서 하나씩 책 제목 출력하기 

```python
print("아래는 장바구니 목록입니다.")
print("")
print("")

for n in cart:            
    print(n.string) 
```

## 4. 전체 코드 다시 보기

```python
from selenium import webdriver
from bs4 import BeautifulSoup

web_crawling = webdriver.Chrome("./chromedriver")
web_crawling.get('http://www.kyobobook.co.kr/login/login.laf?Kc=GNHHNOlogin&orderClick=c03&retURL=http%3A//www.kyobobook.co.kr/index.laf')

web_crawling.find_element_by_name('memid').send_keys('본인아이디')
web_crawling.find_element_by_name('pw').send_keys('본인 비밀번호')

auto_login = web_crawling.find_element_by_css_selector("input.btn_submit[type=submit]")
auto_login.submit()

web_crawling.get('http://order.kyobobook.co.kr/cart/cartListMain')

html = web_crawling.page_source   
soup = BeautifulSoup(html, 'html.parser')  
cart = soup.select('div.book_name > a > span')

print("아래는 장바구니 목록입니다.")
print("")
print("")

for n in cart:            
    print(n.string) 
```
